```python
# Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
```

```python
import numpy as np

x = np.array([0, 1, 2, 3, 4, 5]) 
y = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2,2,1)
plt.plot(x,y)
plt.title('Plot 1')

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2,2,2)
plt.plot(x,y)
plt.title('Plot 2')

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2,2,3)
plt.plot(x,y)
plt.title('Plot 3')

x = np.array([0, 1, 2, 3, 4, 5])
y = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2,2,4)
plt.plot(x,y)
plt.title('Plot 4')

plt.tight_layout()
plt.show()
```

```python
print("A bar plot (or bar chart) is a type of graph that represents categorical data with rectangular bars. The length or height of each bar is proportional to the value it represents.")
print("It is used for :")
print("Comparison of Categories , Visualization of Discrete Data , Clear Representation of Data , Flexibility , Highlighting Trends")
```

```python
company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])
```

```python
# Q5: What is a box plot? Why is it used? Using the following data plot a box plot.
```

```python
print("A box plot (also known as a box-and-whisker plot) is a standardized way of displaying the distribution of data based on a five-number summary: the minimum, first quartile (Q1), median (Q2), third quartile (Q3), and maximum. The plot may also include 'whiskers,' which extend from the quartiles to the minimum and maximum values, and 'outliers,' which are data points that fall outside the range defined by the whiskers.")
print("It is usde for : ")
print("Visualizing Distribution , Identifying Outliers , Comparing Distributions , Understanding Symmetry and Skewness , Handling Large Datasets")
```

```python
box1 = np.random.normal(100, 10, 200)
box2 = np.random.normal(90, 20, 200)
```

```python
plt.boxplot(box1)
```

```python
plt.boxplot(box2)
```
