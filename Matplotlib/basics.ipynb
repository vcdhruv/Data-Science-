```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
x = np.linspace(0,10,200)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.figure(figsize=(6,2)) # for giving figure size
plt.plot(x,y,'--r') # --r means red dotted line , --b means blue dotted line
plt.xlabel("this is my x data") # label x
plt.ylabel("this is my y data") # label y
plt.title("This is my title") # label title
plt.grid() # to show grid lines
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x, y) # scatter
```

```python
x
```

```python
y
```

```python
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c=colors , s = sizes , alpha= 0.7) # colors , sizes and intensity
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("random data")
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)
sizes = 1000 * np.random.rand(50)
plt.scatter(x, y, c="#0f0f0f" , s = sizes , alpha= 0.7 ,marker='v') # colors , sizes , intensity and marker
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("random data")
```
