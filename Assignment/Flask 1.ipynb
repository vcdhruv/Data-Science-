What is App routing in Flask? Why do we use app routes?

```python
print("App routing in Flask refers to the mechanism of mapping URLs to functions (also known as view functions or route handlers) in a Flask web application. This is done using the @app.route() decorator to define URL patterns and associate them with specific functions that handle requests to those URLs")
print("We Use App Routes for URL Mapping , Handling Requests , Separation of Concerns , Dynamic URLs , HTTP Methods.")
```
